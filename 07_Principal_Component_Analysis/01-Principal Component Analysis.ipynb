## PCA Visualization

It is difficult to visualize high dimensional data, we can use PCA to find the first two principal components, and visualize the data in this new, two-dimensional space, with a single scatter-plot. Before doing this though, It is necessary to scale data so that each feature has a single unit variance.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

PCA with Scikit Learn uses a very similar process to other preprocessing functions that come with SciKit Learn. Instantiate a PCA object, find the principal components using the fit method, then apply the rotation and dimensionality reduction by calling transform().

Than specify how many components we want to keep when creating the PCA object.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

transform this data to its first 2 principal components.

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
scaled_data
```

```python
x_pca.shape
```

```python
x_pca
```

Great! We've reduced 30 dimensions to just 2! Let's plot these two dimensions out!

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target'])
plt.xlabel('First Principal Components')
plt.ylabel(('second Principal Components'))
```

Clearly by using these two components it can easily separate these two classes.

## Interpreting the components 

Unfortunately, with this great power of dimensionality reduction, comes the cost of being able to easily understand what these components represent.

The components correspond to combinations of the original features, the components themselves are stored as an attribute of the fitted PCA object:

```python
pca.components_
```

In this numpy matrix array, each row represents a principal component, and each column relates back to the original features. 

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp, cmap='plasma')
```

This heatmap and the color bar basically represent the correlation between the various feature and the principal component itself.

Hopefully this information is useful when dealing with high dimensional data!
